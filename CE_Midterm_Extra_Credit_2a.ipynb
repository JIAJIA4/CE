```python
A = 0xe93ae9c5fc7355d5
B = 0xf43afec7e1684adf

LEN = 8

def data_to_uint8_array(x):
    out = []
    while x > 0:
        out.append(x & 0xFF)
        x = x >> 8
    out = out[::-1]
    return out

def printword(x):
    for h in x:
        print(chr(h), end="")


A = data_to_uint8_array(A)
B = data_to_uint8_array(B)

C = [0] * LEN
D = [0] * LEN
OTP = [0] * LEN
ANS1 = [0] * LEN
ANS2 = [0] * LEN
TEMP_WORD = []
WORD = []

for i in range(LEN):
    C[i] = A[i] ^ B[i]

with open("English_words.txt", "r") as file:
    for line in file:
        TEMP_WORD.append(line.strip())
    
for i in range (len(TEMP_WORD)):
    if len(TEMP_WORD[i]) == LEN:
        WORD.append(TEMP_WORD[i])

for i in range (len(WORD)):
    W = []
    for char in WORD[i]:
        W.append(ord(char))
    WORD[i] = W
print("len = ",len(WORD))
for i in range(len(WORD)):
    for j in range(LEN):
        D[j] = C[j] ^ WORD[i][j]
    for j in range(len(WORD)):
        if WORD[j] == D:
            ANS1 = WORD[j]
            ANS2 = WORD[i]
    if i % 5001 == 5000:
        print("i = ",i, end="  ")
print("")

print("A = ",end=" ")
for i in range(LEN):
    print(hex(A[i])[2:], end=" ")
print("")

print("B = ",end=" ")
for i in range(LEN):
    print(hex(B[i])[2:], end=" ")
print("")

print("C = A^B = ",end=" ")
for i in range(LEN):
    print(hex(C[i])[2:], end=" ")
print("")
            
print("\nANS = ",end="")
printword(ANS1)
print(",  ",end=" ")
printword(ANS2)
print("")

print("OTP1 = ")
for i in range(LEN):
    OTP[i] = ANS1[i] ^ A[i]
for i in range(LEN):
    print(hex(OTP[i])[2:], end=" ")
print("")
for i in range(LEN):
    OTP[i] = ANS2[i] ^ B[i]
for i in range(LEN):
    print(hex(OTP[i])[2:], end=" ")

print("\nOTP2 = ")
for i in range(LEN):
    OTP[i] = ANS1[i] ^ B[i]
for i in range(LEN):
    print(hex(OTP[i])[2:], end=" ")
print("")
for i in range(LEN):
    OTP[i] = ANS2[i] ^ A[i]
for i in range(LEN):
    print(hex(OTP[i])[2:], end=" ")
print("\n")   

print("n           =", hex(ANS1[0]), "=", bin(0x6e))
print("OPT1[0]     = 0x87 =", bin(0x87))
print("0x6e ^ 0x87 = 0xe9 =", bin(0xe9))
```
