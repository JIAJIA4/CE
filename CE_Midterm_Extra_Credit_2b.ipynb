```python
def find_word(all_msg, L, all_otp):
    TEMP_WORD = []
    WORD = []
    with open("English_words.txt", "r") as file:
        for line in file:
            TEMP_WORD.append(line.strip())
        
    for i in range(len(TEMP_WORD)):
        if len(TEMP_WORD[i]) == L:
            WORD.append(TEMP_WORD[i])
            
    for i in range(len(WORD)):
        W = []
        for char in WORD[i]:
            W.append(ord(char))
        WORD[i] = W
    
    for i in range(len(all_msg)):
        for j in range(len(WORD)):
            if WORD[j] == all_msg[i]:
                print("FIND!!! i = ",i, end="  msg = ")
                printword(all_msg[i])
                print("   otp = ",all_otp[i])
```
